```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = "updated_hayf.xlsx"
df_updated = pd.read_excel(file)
df_updated = df_updated.set_index('date', inplace=False)
df_updated.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate counts per artist
artist_counts = df_updated['Artist'].value_counts().reset_index()
artist_counts.columns = ['Artist', 'Counts']

# Sort artists by count
artist_counts = artist_counts.sort_values(by='Counts', ascending=False)

# Create a horizontal bar plot with sorted artists
plt.figure(figsize=(20, 30), facecolor='black')  # Set figure background to black
sns.barplot(data=artist_counts, y='Artist', x='Counts', color='#4F2683')  # Minnesota Vikings purple color

# Axes styling
ax = plt.gca()  # Get current axis
ax.set_facecolor('black')  # Set axes background to black
ax.tick_params(colors='white')  # Set tick colors to white for visibility

# Label, Title, and Tick styling
plt.xlabel('Counts', fontweight='bold', fontsize=16, color='white')
plt.ylabel('Artist', fontweight='bold', fontsize=16, color='white')
plt.title('Number of Concerts by Artist', fontweight='bold', fontsize=16, color='white')
plt.yticks(fontweight='bold', fontsize=16, color='white')  # Make artist names bold and larger, and set to white
plt.xticks(fontweight='bold', fontsize=16, color='white')  # Make x-axis labels bold, larger, and set to white
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,30), facecolor='black')  # Adjust figure size to accommodate all venues, set background to black
venue_counts = df_updated['venue'].value_counts().reset_index()
venue_counts.columns = ['Venue', 'Counts']
venue_counts = venue_counts.sort_values('Counts', ascending=False)
sns.barplot(data=venue_counts, y='Venue', x='Counts', color='#4F2683')  # Minnesota Vikings purple color

# Axes styling
ax = plt.gca()  # Get current axis
ax.set_facecolor('black')  # Set axes background to black
ax.tick_params(colors='white')  # Set tick colors to white for visibility

# Label, Title, and Tick styling
plt.xlabel('Counts', fontweight='bold', fontsize=16, color='white')
plt.ylabel('Venue', fontweight='bold', fontsize=16, color='white')
plt.title('Number of Concerts by Venue', fontweight='bold', fontsize=16, color='white')
plt.yticks(fontweight='bold', fontsize=16, color='white')  # Make venue names bold, larger, and set to white
plt.xticks(fontweight='bold', fontsize=16, color='white')  # Make x-axis labels bold, larger, and set to white

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count concerts per year
year_counts = df_updated['year'].value_counts().sort_index()

# Create a bar plot of concert counts by year
plt.figure(figsize=(20, 10), facecolor='black')  # Set figure background to black
sns.barplot(x=year_counts.index, y=year_counts.values, color='#4F2683')  # Minnesota Vikings purple color

# Axes styling
ax = plt.gca()  # Get current axis
ax.set_facecolor('black')  # Set axes background to black
ax.tick_params(colors='white')  # Set tick colors to white for visibility

# Label, Title, and Tick styling
plt.title('Number of Concerts by Year', fontweight='bold', fontsize=16, color='white')
plt.xlabel('Year', fontweight='bold', fontsize=16, color='white')
plt.ylabel('Number of Concerts', fontweight='bold', fontsize=16, color='white')
plt.xticks(rotation=45, fontweight='bold', fontsize=16, color='white')  # Rotate x-labels for better readability
plt.yticks(fontweight='bold', fontsize=16, color='white')

plt.show()
```
